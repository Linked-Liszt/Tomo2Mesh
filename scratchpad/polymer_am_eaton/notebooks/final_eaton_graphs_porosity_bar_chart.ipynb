```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
sns.set(font_scale = 1.4)
sns.set_style("white")
```

```python
fpath = '/data01/Eaton_Polymer_AM/csv_files/porosity_vals_adj2.csv'
```

```python
df = pd.read_csv(fpath)
df = df.set_index("sample_tag").drop("injmold", axis = 0).reset_index()
df["sample_tag"] = df["sample_tag"].astype(np.uint64)
df = df.sort_values("sample_tag")
df["1/porosity"] = 1.0/(np.array(df["porosity"])*100)

density = np.array([1.085,1.046,1.067,1.058,1.088,1.077,1.073,1.033,1.015,1.064,1.081,1.068])
```

```python
labels = ["$g/cm^3$"]

x = df.groupby("sample_tag")["1/porosity"].mean().index
y = df.groupby("sample_tag")["1/porosity"].mean().values

width = 0.3

sns.set(font_scale=1.2)
sns.set_style(style = "white")
fig, ax = plt.subplots(1,1, figsize = (12,6))
ax2 = ax.twinx()
bar1 = ax.bar(x,y, width, color="black")
_ = ax.set_xticks(np.arange(1,13))
_ = ax.set_xticklabels(np.arange(1,13))
bar2 = ax2.bar(np.array(x)+width,density, width, color = "red")
ax.set_ylabel("1 / Porosity")
ax2.set_ylabel(f"Density ({labels[0]})")
ax.set_xlabel("Sample Number")
ax.set_title("1/Porosity vs. Density")
#ax2.set_ylim([0,1.4])
ax.set_ylim([min(y)-0.002,max(y)+0.002])
ax2.set_ylim([min(density)-0.02,max(density)+0.02])
ax.legend([bar1, bar2],["1/Porosity","Density"])
plt.tight_layout()
```

```python
x = df.groupby("sample_tag")["porosity"].mean().index
y = df.groupby("sample_tag")["porosity"].mean().values

width = 0.3

sns.set(font_scale=1.2)
sns.set_style(style = "white")
fig, ax = plt.subplots(1,1, figsize = (12,6))
ax2 = ax.twinx()
bar1 = ax.bar(x,y, width, color="black")
_ = ax.set_xticks(np.arange(1,13))
_ = ax.set_xticklabels(np.arange(1,13))
bar2 = ax2.bar(np.array(x)+width,density, width, color = "red")
ax.set_ylabel("Porosity")
ax2.set_ylabel(f"Density ({labels[0]})")
ax.set_xlabel("Sample Number")
ax.set_title("Porosity vs. Density")
ax.set_ylim([min(y)-0.002,max(y)+0.002])
ax2.set_ylim([min(density)-0.02,max(density)+0.02])
ax.legend([bar1, bar2],["Porosity","Density"])
plt.tight_layout()
```
